```python
%cd /content/structural-antibody-binding-gnn

import pathlib

path = pathlib.Path("src/gnn/train_gnn.py")
text = path.read_text()

old_line = "    rmse = mean_squared_error(truths_arr, preds_arr, squared=False)"
if old_line in text:
    replacement = (
        "    mse = mean_squared_error(truths_arr, preds_arr)\n"
        "    rmse = float(mse) ** 0.5"
    )
    text = text.replace(old_line, replacement)
    path.write_text(text)
    print("Patched RMSE computation in evaluate().")
else:
    print("Line with squared=False not found; nothing patched.")
```
